```python
# Import setup
import pandas as pd
```

```python
# Create a list containing files for 2023 and 2024 we want to analyse 
files = ['Data/JC-202301-citibike-tripdata.csv',
         'Data/JC-202302-citibike-tripdata.csv',
         'Data/JC-202303-citibike-tripdata.csv',
         'Data/JC-202304-citibike-tripdata.csv',
         'Data/JC-202305-citibike-tripdata.csv',
         'Data/JC-202306-citibike-tripdata.csv',
         'Data/JC-202307-citibike-tripdata.csv',
         'Data/JC-202308-citibike-tripdata.csv',
         'Data/JC-202309-citibike-tripdata.csv',
         'Data/JC-202310-citibike-tripdata.csv',
         'Data/JC-202311-citibike-tripdata.csv',
         'Data/JC-202312-citibike-tripdata.csv',
         'Data/JC-202401-citibike-tripdata.csv',
         'Data/JC-202402-citibike-tripdata.csv']
```

```python
# Create an empty dataframe to store all data
citibike_df = pd.DataFrame()
```

```python
# Loop through the list of files and concatonate all data to the dataframe
for file in files:
    data = pd.read_csv(file) 
    citibike_df = pd.concat([citibike_df, data], axis=0)
```

```python
# Show the dataframe
citibike_df
    
```

```python
# Store the dataframe as CSV file
citibike_df.to_csv('data/2023_2024city_bike_data.csv', index=False)
```

```python
citibike_df.info()
```

```python
# cleaning data as we see different number of rows 

# Check for missing values
print(citibike_df.isnull().sum())
```

```python
# Convert datetime columns to datetime data type
citibike_df['started_at'] = pd.to_datetime(citibike_df['started_at'])
citibike_df['ended_at'] = pd.to_datetime(citibike_df['ended_at'])
```

```python
# Remove duplicate rows
citibike_df.drop_duplicates(inplace=True)
```

```python
# Handling missing values in categorical columns
citibike_df['start_station_name'].fillna('Unknown', inplace=True)
citibike_df['start_station_id'].fillna('Unknown', inplace=True)
citibike_df['end_station_name'].fillna('Unknown', inplace=True)
citibike_df['end_station_id'].fillna('Unknown', inplace=True)

# Handling missing values in numerical columns
citibike_df['end_lat'].fillna(citibike_df['end_lat'].mean(), inplace=True)
citibike_df['end_lng'].fillna(citibike_df['end_lng'].mean(), inplace=True)
```

```python
# Check
citibike_df.info()
```

```python
# Save the cleaned dataset
citibike_df.to_csv('city_bike_dataset.csv', index=False)
```
